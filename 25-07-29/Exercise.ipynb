##Task:
1. Save it as students.csv in Colab.
2.Read it into a pandas DataFrame and print the first 3 rows.  



```python
import pandas as pd
file_path = '/content/drive/My Drive/students.csv'
df = pd.read_csv(file_path)
print(df.head(3))
```

##Task:

1.   Add a new column Average that calculates the mean score for Math, Science, and English for each student.
2. Print the student with the highest average.
  



```python
import pandas as pd
df['average']=df[['Math','Science','English']].mean(axis=1)
print(df)
topper = df.loc[df['average'].idxmax()]
print(f"{topper['Name']} average score {topper['average']:.2f}")
```

##Task:


*   Display all students who have scored more than 80 in Science.
*  Display all students with an average score below 70.



```python
import pandas as pd
above_80 = df[df['Science'] > 80]
print("Students who scored more than 80 in Science:")
print(above_80)
averagebelow = df[df['average'] < 70]
print("\nStudents with an average score below 70:")
print(averagebelow)
```

##Task
Sort the students by English score (descending) and print the top 3.

```python
import pandas as pd
english_score=df.sort_values(by='English' ,ascending=False)
print(english_score.head(3))
```

##Task:
Chart Exercise
1. Create a bar chart showing students’ names on the X-axis and their average
scores on the Y-axis.
2. Create a line chart for all students’ scores in Math, Science, and English (use df.plot() ).

```python
#bar chart
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['average'], color='skyblue')
plt.title('Average Scores of Students')
plt.xlabel(' Name')
plt.ylabel('Average')
plt.ylim(0, 100)
plt.grid(True)
plt.show()

#Line chart
df_plot = df.set_index('Name')[['Math', 'Science', 'English']]
df_plot.plot(kind='line', marker='o', figsize=(8,5))
plt.title("Student Scores")
plt.ylabel("Scores")
plt.ylim(0, 100)
plt.grid(True)
plt.show()

```
