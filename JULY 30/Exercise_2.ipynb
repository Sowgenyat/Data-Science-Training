##Task:

1. Total Sales Per Store, Group the data by Store and calculate total sales for each store.
3. Identify which store had higher total sales.

```python
import pandas as pd
total_sales=df.groupby('Store')['Sales'].sum()
print(total_sales)
high_total_sale=total_sales.idxmax()
print("\nstore with higher total sale\n" ,high_total_sale)
```

##Task:
1. Monthly Sales Trend Group the data by Month and sum up the total sales (Store A + Store B).
Print the result.

```python
import pandas as pd
monthly_sales=df.groupby('Month')['Sales'].sum()
print(monthly_sales)
```

## Task:
1. Chart: Monthly Sales by Store
2. Create a line chart showing monthly sales for both Store A and Store B.
X-axis: Month
Y-axis: Sales
Two lines: Store A and Store B

```python
import pandas as pd
import matplotlib.pyplot as plt
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)
pivot_df = df.pivot(index='Month', columns='Store', values='Sales')
pivot_df = pivot_df.sort_index()
pivot_df.plot(kind='line', marker='o')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

##Bonus
1. Create a bar chart showing total sales per store.
Add data labels on top of each bar.
2. Save this chart as an image ( sales_chart.png ).

```python
import pandas as pd
import matplotlib.pyplot as plt
total = df.groupby('Store')['Sales'].sum()
plt.figure(figsize=(8,5))
plt.bar(total.index, total.values)
plt.title("Total Sales")
plt.xlabel("Store")
plt.ylabel("Sales")
plt.savefig("sales_chart.png")
plt.grid(True)
plt.show()
```
